# Relative Wiener--Hopf factorization trace

```python
import sympy as sp
from IPython.display import Math, display
from symbolic_operator_calculus import build_relative_wiener_hopf_trace, render_relative_wiener_hopf_trace_latex
```

```python
gamma1, gamma2, d = sp.symbols('gamma1 gamma2 d', positive=True, real=True)
trace_12 = build_relative_wiener_hopf_trace(1, 2, gamma1, gamma2, sp.Integer(0), d)
trace_21 = build_relative_wiener_hopf_trace(2, 1, gamma2, gamma1, d, sp.Integer(0))
```

```python
for trace in (trace_12, trace_21):
    for step in render_relative_wiener_hopf_trace_latex(trace).steps:
        display(Math(step.latex))
```
