# 12. Impacto en los 16 términos

Se comprueba estructuralmente cuántas veces aparece cada núcleo \(K_{Q,B}\) dentro de la expansión de Fase O.

```python
phase_o = build_complete_correction_expansion_trace()
def contains_contiguous(factors, subword):
    width = len(subword)
    return any(factors[i:i + width] == subword for i in range(len(factors) - width + 1))
impact = {
    word.identifier: sum(
        contains_contiguous(term.product.factors, word.product.factors)
        for term in phase_o.c3.terms
    )
    for word in analysis.words
}
assert len(phase_o.c3.terms) == 16
assert impact == {"K++": 4, "K+Ghat": 4, "K-+": 4, "K-Ghat": 4}
impact
```
